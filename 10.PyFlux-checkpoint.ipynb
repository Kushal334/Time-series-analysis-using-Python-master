```python
import numpy as np
import pandas as pd
import pyflux as pf
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/sunspot.year.csv')
```

```python
data
```

```python
data.index = data['time'].values
```

```python
plt.figure(figsize=(15,5))
plt.plot(data.index,data['value'])
plt.ylabel('Sunspots')
plt.title('Yearly Sunspot Data');
```

```python
model = pf.ARIMA(data=data, ar=4, ma=4, target='value', family=pf.Normal())
```

```python
x = model.fit("MLE")
```

```python
x.summary()
```

```python
model.plot_z(figsize=(15,5))
```

```python
model.plot_fit(figsize=(15,10))
```

```python
model.plot_predict_is(h=50, figsize=(15,5))
```

```python
model.plot_predict(h=20,past_values=20,figsize=(15,5))
```

```python
import numpy as np
import pandas as pd
import pyflux as pf
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("drivers.csv")
```

```python
data
```

```python
data.index = data['time'];
```

```python
data.index
```

```python
data.loc[(data['time']<1983.05), 'seat_belt'] = 0;
data.loc[(data['time']>=1974.00), 'oil_crisis'] = 1;
data.loc[(data['time']<1974.00), 'oil_crisis'] = 0;
```

```python
plt.figure(figsize=(15,5));
plt.plot(data.index,data['value']);
plt.ylabel('Driver Deaths');
plt.title('Deaths of Car Drivers in Great Britain 1969-84');
plt.plot();
```

```python
model = pf.ARIMAX(data=data, formula='value~1+seat_belt+oil_crisis',ar=1, ma=1, family=pf.Normal())
```

```python
x = model.fit("MLE")
x.summary()
```

```python
model.plot_fit(figsize=(15,10))
```
