```python
your_time_series = np.random.rand(10000)
window_size = 50
```

```python
matrix_profile = stumpy.stump(your_time_series, m=window_size)

subseq_len = 50
```

```python
correct_arc_curve, regime_locations = stumpy.fluss(matrix_profile[:, 1],
                                                   L=subseq_len,
                                                   n_regimes=2,
                                                   excl_factor=1
                                                  )
```
